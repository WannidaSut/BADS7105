```python
df = pd.read_csv('Customer Survey.csv')
df.head()
```

```python
df.replace({'เคย':1,'ไม่เคย':0},inplace=True)
df.fillna(1,inplace=True)
df.head(5)
```

```python
df.drop(labels=df.columns[0:2].tolist(), axis=1, inplace=True)
```

```python
df
```

```python
data_ibs = pd.DataFrame(index=df.columns,columns=df.columns)
data_ibs.head(5)
```

```python
for i in range(0,len(data_ibs.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(data_ibs.columns)) :
      # Fill in placeholder with cosine similarities
      data_ibs.iloc[i,j] = 1-cosine(df.iloc[:,i],df.iloc[:,j])
      
data_ibs
```

```python
links = data_ibs.stack().reset_index()
links.columns = ['edge1', 'edge2','value']
links
```

```python
links_filtered=links.loc[ (links['value'] > 0.7) & (links['edge1'] != links['edge2']) ]
links_filtered
```

```python
net = Network(notebook=True)
G=nx.from_pandas_edgelist(links_filtered, 'edge1', 'edge2', 'value')
net.from_nx(G)
```

```python
net.show("example.html")
```
